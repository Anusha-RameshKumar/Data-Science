```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("E:\data science\Preprocessing-Project\data\creditcard.csv")
df.head()
df.info()
df.describe().T
```

```python
plt.figure(figsize=(7,4))
sns.boxplot(y=df['V8'])
plt.title("Boxplot")
plt.show()
```

```python
plt.pie([20,30,50,70,40],colors=['Yellow','Red','Purple','Brown','Orange'],labels=['Clothing','Books','Sports','Electronics','Beauty items'])
plt.title("Pie Plot")
plt.title("Products")
plt.ylabel("")
plt.show()
```

```python
sample=df.head(10)
plt.figure(figsize=(12,6))
plt.plot(sample['V1'], sample['V28'], marker='o')
plt.xlabel("Order Date")
plt.ylabel("Total Sales")
plt.title("Total Sales Over Time")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
sample=df.sample(30, random_state=1)
plt.figure(figsize=(6,6))
sns.scatterplot(data=sample, x='V1', y='V28')
plt.xlabel("V1")
plt.ylabel('V28')
plt.title("User_ID vs 'Total_Votes'")
plt.grid(True)
plt.show()
```
